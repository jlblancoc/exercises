```python
from mass_spring_damper import mass, spring, damper
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.animation
%matplotlib inline
        
fig, ax = plt.subplots()
ax.axis([0,16,-2.5,2.5])
ax.set_xlabel('Length of a spring-damper-mass system')
ax.set_title('Length of a spring-damper-mass system')
ax.grid(True)
l1, = ax.plot([],[],'b')
l2, = ax.plot([],[],'b')
l3, = ax.plot([],[],'b')
l4, = ax.plot([],[],'b')

NF = 50
sp = spring(8.0)
sp.locate_spring(0.,-1.,0.)
da = damper(8.0)
da.locate_damper(0.,+1.,0.)
ms = mass(6,3)
ms.locate_mass(11.,0.,0.)

def animate(i):
    x = 1*np.sin(2*np.pi*i/NF)
    sp.deform(x)
    da.deform(x)
    ms.deform(x)
    l1.set_data(sp.global_x,sp.global_y)
    l2.set_data(da.global_rod_x,da.global_rod_y)
    l3.set_data(da.global_cyl_x,da.global_cyl_y)
    l4.set_data(ms.global_x,ms.global_y)

ani = matplotlib.animation.FuncAnimation(fig, animate, frames = NF)

from IPython.display import HTML
HTML(ani.to_jshtml())
```
